```python
import random 
def sample(fileName, K):
    reservoir = []
    N=0 #Î±ÏÎ¹Î¸Î¼ÏŒÏ‚ Î´Î¹Î±Î²Î±ÏƒÎ¼Î­Î½Ï‰Î½ Î±Î½Ï„Î¹ÎºÎµÎ¹Î¼Î­Î½Ï‰Î½
    file=open(fileName, 'r')
    
    for line in file:
        N+=1
        probability=K/N #Ï€Î¹Î¸Î±Î½ÏŒÏ„Î·Ï„Î± ÎµÎ¼Ï†Î¬Î½Î¹ÏƒÎ·Ï‚ ÏƒÏ„Î¿ Î´ÎµÎ¯Î³Î¼Î±
        
        #ÎµÏ€Î¹Î»Î¿Î³Î® Ï„Î¿Ï… Î±Î½Ï„Î¹ÎºÎµÎ¹Î¼Î­Î½Î¿Ï… Î¼Îµ Ï€Î¹Î¸Î±Î½ÏŒÏ„Î·Ï„Î± probability
        if random.randint(0,1) < probability: 
            #Î±Î½ Ï„Î¿ Î´ÎµÎ¹Î³Î¼Î± Î´ÎµÎ½ ÎµÏ‡ÎµÎ¹ Ï†Ï„Î±ÏƒÎµÎ¹ Ï„Î¿ Î¼Î­Î³ÎµÎ¸Î¿Ï‚ Îš Ï€ÏÎ¿ÏƒÎ¸ÎµÏ„Ï‰ Ï„Î¿ Î±Î½Ï„Î¹ÎºÎµÎ¯Î¼ÎµÎ½Î¿ ÏƒÏ„Î¿ reservoir
            if len(reservoir)<K:
                reservoir.append(line.strip())
            else: 
                #Î±Î½ Ï„Î¿ Î´ÎµÎ¯Î³Î¼Î± Î­Ï‡ÎµÎ¹ Ï†Ï„Î¬ÏƒÎµÎ¹ Ï„Î¿ Î¼Î­Î³ÎµÎ¸Î¿Ï‚ Îš, Ï„Î¿ Î±Î½Ï„Î¹ÎºÎµÎ¯Î¼ÎµÎ½Î¿ Î±Î½Ï„Î¹ÎºÎ±Î¸Î¹ÏƒÏ„Î¬ Ï…Ï€Î¬ÏÏ‡Î¿Î½ ÏƒÏ„Î¿ reservoir
                replace_index=random.randint(0,K-1)
                reservoir[replace_index]=line.strip()
    return reservoir 
                
```
