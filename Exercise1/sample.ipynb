```python
import random 
def sample(fileName, K):
    reservoir = []
    N=0 #αριθμός διαβασμένων αντικειμένων
    file=open(fileName, 'r')
    
    for line in file:
        N+=1
        probability=K/N #πιθανότητα εμφάνισης στο δείγμα
        
        #επιλογή του αντικειμένου με πιθανότητα probability
        if random.randint(0,1) < probability: 
            #αν το δειγμα δεν εχει φτασει το μέγεθος Κ προσθετω το αντικείμενο στο reservoir
            if len(reservoir)<K:
                reservoir.append(line.strip())
            else: 
                #αν το δείγμα έχει φτάσει το μέγεθος Κ, το αντικείμενο αντικαθιστά υπάρχον στο reservoir
                replace_index=random.randint(0,K-1)
                reservoir[replace_index]=line.strip()
    return reservoir 
                
```

```python
fileName='input.txt'
K=10
res=sample(fileName,K)
for line in res:
    print(line)
```

ΠΕΡΙΓΡΑΦΗ ΑΛΓΟΡΙΘΜΟΥ
Με τον αλγόριθμο Reservoir Sampling εκτελούμε δειγματοληψία από ένα ρεύμα δεδομένων, χωρίς να γνωρίζουμε εκ των προτέρων το μέγεθος του ρεύματος. Αρχικά, δημιουργούμε έναν πίνακα δειγματοληψίας για την αποθήκευση των αντικειμένων που θα επιλέξουμε, τα οποία έχουν πιθανότητα επιλογής έναν αριθμό που θα ορίσουμε εμείς δια τον αύξοντα αριθμό του αντικειμένου που διαβάζουμε εκείνη τη στιγμή στο ρεύμα. Στη συνέχεια διαβάζουμε τα αντικείμενα ένα – ένα από το ρεύμα και για το καθένα υπολογίζουμε την παραπάνω πιθανότητα, με την οποία κρίνουμε αν το αντικείμενο αυτό θα μπει στο ρεζερβουάρ με το τελικό δείγμα. Αν η πιθανότητα αυτή ενός αντικειμένου είναι μικρότερη του 1, τότε το αντικείμενο αυτό δεν είναι τόσο πιθανό να μπει στο ρεζερβουάρ. Τελικώς, το δείγμα αποτελείται από τα αντικείμενα που επιλέχθηκαν στον πίνακα δειγματοληψίας. Συμπερασματικά, ο αλγόριθμος αυτός εξασφαλίζει ότι κάθε αντικείμενο έχει ίση πιθανότητα επιλογής. 

ΑΠΟΔΕΙΞΗ ΑΛΓΟΡΙΘΜΟΥ
Για να αποδείξουμε ότι ο παραπάνω τροποποιημένος αλγόριθμος παράγει ένα ομοιόμορφο τυχαίο δείγμα θα χρησιμοποιήσουμε τη μέθοδο της μαθηματικής επαγωγής. Σε αυτόν τον αλγόριθμό κάθε στοιχείο 𝑖 έχει πιθανότητα Κ/𝑖 να επιλεγεί από το ρεύμα. Θεωρούμε P(𝑖) την πιθανότητα το 𝑖-οστό στοιχείο να επιλεγεί κατά το τέλος του αλγορίθμου, όταν δηλαδή έχουν διαβαστεί όλα τα στοιχεία. 
Για 𝑖=1 η πιθανότητα επιλογής του πρώτου στοιχείου είναι P(𝑖)=K/1=K. Στη συνέχεια, θα δείξουμε ότι αν ισχύει P(𝑖)=K/𝑖, τότε ισχύει και P(𝑖+1)=K/𝑖+1 για  1≤i≤N-1. Στη περίπτωση που 𝑖+1 είναι μικρότερο ή ίσο του τρέχοντος μεγέθους του δείγματος (t), τότε η πιθανότητα επιλογής του 𝑖+1-οστού στοιχείου είναι P(𝑖+1)=K/𝑖+1. Στη περίπτωση που 𝑖+1 είναι μεγαλύτερο ή ίσο του τρέχοντος μεγέθους του δείγματος (t), τότε η πιθανότητα επιλογής του 𝑖+1-οστού στοιχείου είναι P(𝑖+1)=K/t.                                Σε κάθε περίπτωση, η πιθανότητα επιλογής του 𝑖 +1-οστού στοιχείου είναι ακριβώς η πιθανότητα που θα θέλαμε, δηλαδή Κ/𝑖+1 
Συνεπώς, από επαγωγή, έχουμε ότι για κάθε 𝑖=1,2,…,Ν, ισχύει P(𝑖)=K/𝑖, δηλαδή P(𝑖)=K/Ν.
