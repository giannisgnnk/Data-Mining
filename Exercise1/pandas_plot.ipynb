```python
import pandas as pd #pandas library 
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv') #φόρτωση του data.csv

#Γραφική παράσταση του Β ως προς το Α 
plt.figure(figsize=(8, 6))
plt.plot(df['A'], df['B'], marker='o', linestyle='-', color='b', label='Στήλη B')
plt.xlabel('Στήλη A')
plt.ylabel('Στήλη B')
plt.title('Γραφική Παράσταση Στήλης B ως προς Στήλη A')
plt.legend()
plt.show()

#Γραφικ΄η παρ΄άσταση του Γ ως προς το Α 
plt.figure(figsize=(8, 6))
plt.plot(df['A'], df['C'], marker='o', linestyle='-', color='r', label='Στήλη C')
plt.xlabel('Στήλη A')
plt.ylabel('Στήλη C')
plt.title('Γραφική Παράσταση Στήλης C ως προς Στήλη A')
plt.legend()
plt.show()
```

Αξίζει να σημειωθεί ότι η γραφική παράσταση του C ως προς το Α αυξάνεται με μεγαλύτερο ρυθμό από ότι αυτή του Β ως προς το Α.
