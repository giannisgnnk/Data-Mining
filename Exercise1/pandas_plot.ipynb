```python
import pandas as pd #pandas library 
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv') #  data.csv

#        
plt.figure(figsize=(8, 6))
plt.plot(df['A'], df['B'], marker='o', linestyle='-', color='b', label=' B')
plt.xlabel(' A')
plt.ylabel(' B')
plt.title('   B    A')
plt.legend()
plt.show()

#        
plt.figure(figsize=(8, 6))
plt.plot(df['A'], df['C'], marker='o', linestyle='-', color='r', label=' C')
plt.xlabel(' A')
plt.ylabel(' C')
plt.title('   C    A')
plt.legend()
plt.show()
```
