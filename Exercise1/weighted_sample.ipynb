ΕΞΗΓΗΣΗ ΑΛΓΟΡΙΘΜΟΥ
Προκειμένου να υλοποιήσουμε τον τροποποιημένο αλγόριθμο του Reservoir Sampling για τη σταθμισμένη δειγματοληψία ενός αντικειμένου από ενα ρεύμα, δημιουργούμε και αρχικοποιούμε στο μηδέν εναν μετρητή που κρατάει το συνολικό βάρος των αντικειμενων που έχουν διαβαστεί από το ρεύμα αλλά και έναν πίνακα δειγματοληψίας για την αποθήκευση των αντικειμένων που θα επιλέξουμε αργότερα. Στη συνέχεια, αφού διαβάζουμε γραμμή-γραμμή το αρχείο κρατάμε σε μια μεταβλητή το βάρος του εκάστοτε αντικειμένου, το οποίο βάρος περιέχεται μέσα στο αρχείο και σε κάθε επανάληψη όπου διαβάζουμε επόμενο αντικείμενο, προσθέτουμε το βάρος του στον μετρητή με το συνολικό βάρος. Έπειτα, ορίζουμε την πιθανότητα επιλογής του αντικειμένου, η οποία αποτελείται από το βάρος του αντικειμένου που διαβάζουμε εκείνη τη στιγμή δια τον μετρητή με το συνολικό βάρος των αντικειμένων που έχουν διαβαστεί μέχρι τότε. Αν η παραπάνω πιθανότητα ενός αντικειμένου είναι μικρότερη του 1, τότε το αντικείμενο αυτό δεν είναι τόσο πιθανό να μπει στον πίνακα δειγματοληψίας, δηλαδή στο ρεζερβουάρ. Με αυτόν τον τρόπο, ο αλγόριθμος αυτός δουλεύει με σταθερό χώρο μνήμης, ανεξάρτητο του πλήθους των αντικειμένων στο ρεύμα.
